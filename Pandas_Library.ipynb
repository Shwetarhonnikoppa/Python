# **Customizing line plot**

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x = np.arange(0,10,1)
y = 2*x+5
fig=plt.figure(figsize=(10,5))
plt.plot(x,y,linewidth=2.0,color='k',marker ='*',alpha=1.0,linestyle='--',markerfacecolor='g')
plt.title("Line Plot Demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(['line1'],loc='best')
plt.grid(True)
plt.show()
```

# **How to create bar chart**

```python
#from ast import NameConstant
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#creating dataset
data = {'apples':20,'Mangoes':15,'Lemon':30,'Oranges':10}
names = list(data.keys())
values = list(data.values())
#create bar plot
plt.bar(names,values,color='green') #  barh - horizontal bars
plt.title("Bar Plot Demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

# **Customizing Scatter plot**

```python
import matplotlib.pyplot as plt
%matplotlib inline
#creating dataset
a = [10,20,30,40,50,60,70,80]
b = [5,3,2,5,6,1,4,2]
x = [1,2,3,4,3,2,1,2]
#Creating the scatter plot
plt.scatter(a,b, c='r',s=100,edgecolors='y',marker='o',alpha=0.7)
plt.scatter(a,x, c='k',s=100,edgecolors='y',marker='d',alpha=1)
plt.legend(['b','x'],loc='best')
plt.title("Scatter Plot Demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.savefig('Scatter plot.png')
plt.show()
```

# **How to create a Histogram**

```python
import matplotlib.pyplot as plt
%matplotlib inline
age = [10,12,16,19,11,20,26,28,30,38,35,34,45,60,68,64,62,70,78,75,7985,94,95]
plt.hist(age, bins=[0,20,40,60,80,100])
plt.title("Histogram Demo")
plt.xlabel("Range of values")
plt.ylabel("Frequency of value")
plt.grid(True)
plt.savefig("Histogram plot.pdf")
plt.show()
```

# **How to create Pie chart**

```python
import matplotlib.pyplot as plt
%matplotlib inline
#preparing the dataset
animal = ['Dog','Cat','Wolf','Lion']
count = [50,45,60,80]
#creating pie chart
plt.pie(count,labels=animal,autopct='%1.1f%%',shadow=True,startangle=90,explode=(0,0.1,0,0))
plt.title("Pie Chart Demo")
plt.show()
```

## **How to create Donut chart**

```python
import matplotlib.pyplot as plt
%matplotlib inline
#define the dataset
group_names = ["GroupA","GroupB","GroupC"]
group_size = [20,30,50]
size_center = [10]
a,b,c = [plt.cm.Blues,plt.cm.Reds,plt.cm.Greens]
#Creating the doughnut chart
pie1 = plt.pie(group_size,labels=group_names,autopct='%1.1f%%',radius=1.5,shadow=True,startangle=90,colors=[a(0.5),b(1.0),c(0.5)])
pie2 = plt.pie(size_center,radius=1.0,colors='w')
plt.show()
```
